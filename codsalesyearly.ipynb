```python
dfCallOfDuty = df[df["Name"].str.contains("Call of Duty", na = False)]
#dfCallOfDuty = dfCallOfDuty[dfCallOfDuty["Platform"].isin(["PS", "PS2", "PS3", "PS4"])]
len(dfCallOfDuty.index)
dfCallOfDuty.head(20)

dfCallOfDutyGrouped = dfCallOfDuty.groupby(by = "Year_of_Release").sum()

dfCallOfDutyGrouped.to_csv("D:/Jupyter Notebooks/VgChartzCallOfDuty/codgroupedyear.csv")
dfCallOfDuty = pd.read_csv("D:/Jupyter Notebooks/VgChartzCallOfDuty/codgroupedyear.csv")
```

```python
fig, ax = plt.subplots()

ax.bar(dfCallOfDuty["Year_of_Release"], dfCallOfDuty["Global_Sales"])

plt.xticks(fontsize = 12)
plt.figure(figsize = (19.2, 10.8))

ax.set_ylabel("Sales (in millions")
ax.set_title("Call of Duty Franchise Yearly Sales")
ax.annotate("2016 sales through December 22nd", xy = (2016, 7.5), xytext = (2018, 15), arrowprops = dict(facecolor = "black"))
plt.show()
```
